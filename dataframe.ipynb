# 데이터 타입 변경
# .astype(): 변경하고자 하는 column 이름과 데이터 타입을 딕셔너리로 받음

```python
def remove_comma(x):
  return x.replace(",","")

data = [    
      ['1,000','1,100','1,510'],
      ['1,410','1,420','1,790'],
      ['850','900','1,185']
]

columns = ['03/02', '03/03', '03/04']

df = pd.DataFrame(data=data,columns=columns)
df = df.applymap(remove_comma)
df = df.astype({'03/02':np.int64, '03/03':np.int64, '03/04':np.int64})
print(df.dtypes)
```

# column의 문자열 다루기
# str

```python
def remove_comma(x):
  return x.replace(",","")

data = [    
      ['1,000','1,100','1,510'],
      ['1,410','1,420','1,790'],
      ['850','900','1,185']
]

columns = ['03/02', '03/03', '03/04']

df = pd.DataFrame(data=data,columns=columns)

df['03/02'] = df['03/02'].str.replace(',','').astype(np.int64)
df['03/03'] = df['03/03'].str[:-1]
print(df)
```

# 조건에 맞춰 필터링
# Query: 큰 따옴표로 전체 쿼리를 정의하고 작은 따옴표로 문자열

```python
data = [
        {'cd':'A060310', 'nm':'3S', 'open':2920, 'close':2800},
        {'cd':'A095570', 'nm':'AJ네트웍스', 'open':1920, 'close':1900},
        {'cd':'A006840', 'nm':'AK홀딩스', 'open':2020, 'close':2010},
        {'cd':'A054620', 'nm':'APS홀딩스', 'open':3120, 'close':3200}
]

df = pd.DataFrame(data=data)
df = df.set_index('cd')

print(df.query("nm=='3S'"))
print(df.query("open>close"))
print(df.query("nm in ['3S','AK홀딩스']"))

name = 'AJ네트웍스'
print(df.query("nm==@name"))
```

# filter: items로 선택할 컬럼의 이름을 지정 가능(axis=0이면 인덱스에 대해 검색), regex로 컬럼 문자열의 일부를 필터링 가능(axis=0이면 인덱스에 대해 검색)

```python
data = [
        {'cd':'A060310', 'nm':'3S', 'open':2920, 'close':2800},
        {'cd':'A095570', 'nm':'AJ네트웍스', 'open':1920, 'close':1900},
        {'cd':'A006840', 'nm':'AK홀딩스', 'open':2020, 'close':2010},
        {'cd':'A054620', 'nm':'APS홀딩스', 'open':3120, 'close':3200}
]

df = pd.DataFrame(data=data)
df = df.set_index('cd')

df.filter(items=['nm'])
df.filter(items=['A060310'], axis=0)
df.filter(regex='^c')
df.filter(regex='0$', axis=0)
```

# 정렬 및 순위
# .sort_values('column 이름') / .sort_values(by=column 이름): 오름차순 정렬(내림차순은  ascending=False)
# .rank():순위가 저장된 시리즈 객체를 리턴

```python
data = [
        ['037730', '3R', 1510],
        ['036360', '3SOFT', 1790],
        ['005670', 'ACTS', 1185]
]

columns = ['종목코드', '종목명', '현재가']
df = pd.DataFrame(data=data,columns=columns)
df.set_index('종목코드',inplace=True)
df2 = df.sort_values('현재가',ascending=False)
df2['순위'] = df2['현재가'].rank()
df2.sort_values(by='순위', inplace=True)
print(df2)
```

# .groupby('column 이름'): 그냥은 출력 안 됨
# .get_group('group 이름'): 특정 group만 가져오기
# .agg({'column 이름':[적용하고자 하는 함수들],..}): column마다 다른 함수 적용

```python
data = [
        ['2차전지(생산)', 'SK이노베이션', 10.19, 1.29],
        ['해운', '팬오션', 21.33, 0.95],
        ['시스템반도체', '티엘아이', 35.97, 1.12],
        ['해운', 'HMM', 21.52, 3.20],
        ['시스템반도체',  '아이에이', 37.32, 3.55],
        ['2차전지(생산)', 'LG화학', 83.06, 3.75]
]

columns = ['테마', '종목명', 'PER', 'PBR']
df = pd.DataFrame(data=data,columns=columns)

gb = df.groupby('테마')[['테마', 'PER', 'PBR']].agg({'PER':[max,min],'PBR':[np.std,np.var]})
gb
```

# 무지성 이어붙이기가 아닌 특정 column 값을 기준으로 병합
# .merge(left=df1, right=df2, how='inner'/'outer'/'left'/'right'(기준이 되는 데이터프레임), on=합칠 기준 column명, left_on=, right_on=(column명이 다를 경우 각자 column명 적기))

```python
data = [
        ['전기전자',  '005930', '삼성전자', 74400],
        ['화학', ' 051910', 'LG화학', 896000],
        ['서비스업', '035720', '카카오', 121500]
]
columns = ['업종', '종목코드', '종목명', '현재가']
df1=pd.DataFrame(data=data,columns=columns)

data = [
        ['은행', 2.92],
        ['보험',  0.37],
        ['화학', 0.06],
        ['전기전자', -2.43]
]
columns = ['업종', '등락률']
df2=pd.DataFrame(data=data,columns=columns)

df = pd.merge(left=df1,right=df2,how='right',on='업종')
df
```

# 무작정 이어붙이기가 아닌 특정 index를 기준으로 병합
# .join(other=df, how='left'/'right'/'inner'/'outer'(기본은 'left'), on=column 명(만약 index로 설정이 안 되어있다면)): 데이터프레임 반환

```python
data = [
        ['전기전자',  '005930', '삼성전자', 74400],
        ['화학', ' 051910', 'LG화학', 896000],
        ['서비스업', '035720', '카카오', 121500]
]
columns = ['업종', '종목코드', '종목명', '현재가']
df1=pd.DataFrame(data=data,columns=columns)
df1.set_index('업종',inplace=True)

data = [
        ['은행', 2.92],
        ['보험',  0.37],
        ['화학', 0.06],
        ['전기전자', -2.43]
]
columns = ['항목', '등락률']
df2=pd.DataFrame(data=data,columns=columns)
df2.set_index('항목',inplace=True)

df1 = df1.join(other=df2)
df1
```

# stack/unstack
# .stack(level=(기본은 높은 레벨)): column의 항목을 index로 옮기기
# .unstack: index의 항목을 column으로 옮기기

```python
data = [
        [1000,900,800,700],
        [1200,1400,900,800]
]

level_0 = ['영업이익', '당기순이익']
level_1 = ['컨센서스', '잠정치']
columns = pd.MultiIndex.from_product([level_0,level_1])

df = pd.DataFrame(data=data,columns=columns)
df
```

```python
df.stack(level=0)
```

```python
df.stack().stack().to_frame()
```

```python
df.stack().unstack()
```

```python
data = [
        [1000,1100,900,1200,1300],
        [800,2000,1700,1500,1800]
]
index = ['자본금', '부채']
columns = ['2020/03', '2020/06', '2020/09', '2021/03', '2021/06']
df = pd.DataFrame(data=data,index=index,columns=columns)
df
```

```python
df_stacked = df.stack().reset_index()
df_stacked
```

```python
temp = df_stacked['level_1'].str.split('/')
temp
```

```python
df_split = pd.DataFrame(list(temp))
df_split
```

```python
df_merged = pd.concat([df_stacked, df_split], axis=1)
df_merged.columns = ['계정', '년월', '금액', '연도', '월']
df_merged
```

```python
df_group = df_merged.groupby(['계정','연도']).sum()
df_group
```

```python
df_unstack = df_group.unstack()
df_unstack
```

```python
result = df_unstack['금액']
result.columns.name = ''
result.index.name = ''
result
```

# .melt: 컬럼의 수가 많은 데이터프레임을 세로로 긴 데이터포맷으로 재구조화
